```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
a=np.arange(20)
plt.plot(a)
plt.show()
```

```python
x=np.arange(1,6)
y=2*x+1
plt.title("Matplotlib demo")
plt.xlabel("x axis caption")
plt.ylabel("y axis caption")
plt.plot(x,y)
plt.show()
```

```python
x=np.arange(1,11)
y=3*x+7
plt.title("scatter plotting")
plt.plot(x,y,"ob")
plt.show()
```

```python
x=np.arange(0,2*np.pi,0.1)
y=np.sin(x)
plt.title("sin wave graph")
plt.plot(x,y)
plt.show()
```

```python
x=[2,4,8,6]
y=[5,7,2,1]
x2=[1,5,3,9]
y2=[3,5,7,9]
plt.bar(x,y,align='center')
plt.bar(x2,y2,color='g',align='center')
plt.title("Bar Graph")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```
