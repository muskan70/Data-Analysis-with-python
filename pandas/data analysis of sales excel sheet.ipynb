```python
from xlrd import open_workbook
import numpy as np
wb=open_workbook('sales.xlsx')
sheet=wb.sheet_by_name('Sheet1')
row=sheet.nrows
col=sheet.ncols
print("no.of records:",row)
print("no.of attributes:",col)
```

```python
unique_vendor_names=set()
for i in range(0,row):
    unique_vendor_names.add(sheet.cell(i,col-1).value)
    
print("no.of unique vendors:",len(unique_vendor_names))  
print("unique vendor list:")
for i in unique_vendor_names:
    print(i)
```

```python
d={}
for i in unique_vendor_names:
    d[i]=0
for i in range(0,row):
    x=sheet.cell(i,col-1).value
    d[x]+=1
    
x=[]
y=[]
for i in d:
    print(i," ",d[i])
    x.append(i)
    y.append(d[i])
from matplotlib import pyplot as plt
plt.bar(x,y)
plt.show()
```

```python
print("mean of vendor works",np.mean(y))
plt.scatter(x,y)
plt.show()
plt.plot(x,y)
plt.show()
```
